```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('Five_Year_Data.csv')
```

```python
df.head()
```

```python
df2= df.pivot_table(index= 'country', columns= 'year', values= 'lifeExp')
```

```python
print(df2)
```

```python
cols= df2.columns
print(cols)
df3= df2
for col in cols:
    df3[col]= df3[col].apply(np.int64)
print(df3)
```

```python
plt.subplots(figsize= (5, 40))
sns.heatmap(df3, annot= True, fmt= 'd').get_figure().savefig('Assignment_output.png')
```
